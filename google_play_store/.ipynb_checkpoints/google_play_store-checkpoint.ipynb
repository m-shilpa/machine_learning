```python
# dropping columns
gp = gp.drop(['Category','Last Updated','Current Ver'],axis=1)
```

```python
# removing , and + from Installs column
install =[]
for i in gp['Installs']:
    j = i.replace("+","")
    install.append(j.replace(",",""))
```

```python
# concatinating the new Installs column
install = pd.DataFrame(install,columns=['Install'])
gp = pd.concat([gp,install],axis=1)
```

```python
# dropping the old installs column
gp = gp.drop(["Installs"],axis=1)
gp.head()
```

```python
gp = gp.fillna(gp.mean())
```
