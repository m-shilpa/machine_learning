```python
def hypothesis(a,x1,b,x2,c):
    return (1/(1+np.exp(-(a*x1+b*x2+c))))
```

```python
def error(a,x1,b,x2,c,y):
    e = 0
    m = len(x1)
    for i in range(m):
        e += -(y[i]*np.log(hypothesis(a,x1[i],b,x2[i],c)))-((1-y[i])*np.log(1-hypothesis(a,x1[i],b,x2[i],c)))
        
    return (1/m)*e
```

```python
def step_gradient(a,x1,b,x2,c,y,learning_rate):
    grad_a = 0
    grad_b = 0
    grad_c = 0
    m = len(x1)
    
    for i in range(m):
        grad_a += (hypothesis(a,x1[i],b,x2[i],c)-y[i])*x1[i]
        grad_b += (hypothesis(a,x1[i],b,x2[i],c)-y[i])*x2[i]
        grad_c += (hypothesis(a,x1[i],b,x2[i],c)-y[i])
        
    a = a - (grad_a * learning_rate)
    b = b - (grad_b * learning_rate)
    c = c - (grad_c * learning_rate)
    
    return a,b,c
```

```python
def descend(init_a,x1,init_b,x2,init_c,y,learning_rate,iterations):
    a = init_a
    b = init_b
    c = init_c
    
    for i in range(iterations):
        e = error(a,x1,b,x2,c,y)
        
        if i%1000 == 0:
            print(f"error: {e}, a:{a}, b:{b}")
            
        a,b,c = step_gradient(a,x1,b,x2,c,y,learning_rate)
        
    return a,b,c
```

```python
a =0
b=0
c =0
learning_rate = 0.01
iterations =10000
```

```python
final_a,final_b,final_c = descend(a,x1,b,x2,c,y,learning_rate,iterations)
```

```python
def accuracy(a,x1,b,x2,c,y):
    correct = 0
    m = len(x1)
    
    for i in range(m):
        pred = hypothesis(a,x1[i],b,x2[i],c)
        
        if pred > 0.5:
            z=1
        else:
            z = 0                 # 0 -benign
            
        if y[i] == z:
            correct += 1
            
    print(f"Accuracy: {correct/len(y)}")    
```
